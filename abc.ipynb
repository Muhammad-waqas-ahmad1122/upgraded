```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#define 4 subjects against which student got marks
subjects = ['Physics','Chemistry', 'Math', 'Biology']
```

```python
#three students obtained these marks in above mentioned subjects
y1 = [90, 87, 95, 50]
y2 = [100, 80, 86, 90]
y3 = [70,60,50,20]
```

```python
#here we can plot a graph and modify it with following functions
plt.plot(subjects, y1, Label = 'Smith', linestyle = 'dotted', color= 'black', linewidth = 2, marker = 'o')
plt.plot(subjects, y2, label = 'John', linestyle = 'dashed', color = 'blue', linewidth = 3, marker = '>', markersize = 9)
plt.plot(subjects,y3, label = 'Anna', linestyle = 'dashdot', color = 'orange', linewidth = 4, marker = '<', markersize = 9)

#to plot legends on graph
plt.legend()

#to save file seperatly as in required format
plt.savefig('Yahoo1.png')
```
